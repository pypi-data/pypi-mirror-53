## Screening

In this notebook, I will make intervals for coefficients after a marginal screening.

For visualization purposes, we'll screen a continuous process.

My first covariance will be roughish -- $AR(1)$ or a discretized Ornstein-Uhlenbeck:
$$
Cov(Z_t,Z_s) = e^{-\gamma|t-s|}.
$$

```python
def AR1(n,rho,sigma=1):
    """
    Simulate an AR(1) time series for $n$ steps.
    
    The initial value is drawn from the stationary distribution.
    """
    
    var = sigma**2 / (1 - rho**2)
    Z = np.zeros(n)
    Z[0] = np.random.standard_normal() * np.sqrt(var)
    for i in range(n-1):
        Z[i+1] = rho * Z[i] + sigma * np.random.standard_normal()
    rho_pow = rho**np.arange(n)
    cov = np.identity(n) / 2.
    for i in range(1,n):
        cov += np.diag(rho**i*np.ones(n-i),i)
    cov += cov.T
    return Z, cov*sigma**2 / (1 - rho**2)
```

```python
n, rho, threshold = 400, .95, 2.5
Z, covZ = AR1(n, rho)
```

```python
f = plt.figure(figsize=(12,7))
ax = plt.gca()

plt.plot(Z, color='gray', linewidth=2)
screenZ = positive_screen(Z, covZ, threshold * np.sqrt(covZ[0,0]))
mu = 0
for i, eta, center, interval in screenZ.intervals:
    lower, upper = interval
    truth = (eta * mu).sum()
    if plot:
        plt.scatter(i, truth, s=30)
    if truth > lower and truth < upper:
        covered = True
    else:
        covered = False
    ylim = ax.get_ylim()
    if plot:
        if not covered:
            plt.plot([i,i], [lower, upper], linewidth=5, color='red', alpha=0.4)
        else:
            plt.plot([i,i], [lower, upper], linewidth=5, color='green', alpha=0.4)
    #ax.set_ylim(ylim)
ax.set_xlim([-1,n+1])
```

Ouch. These are big. Maybe we should form intervals for the average of the mean over the connected components of the selected regions? Also, we should probably coarsen the partition over this connected component if possible.

## Adding a signal

Let's add a signal in the RKHS and see if things look a little better. I hope so...

```python
mu = 3 * covZ[n/3] 
mu /= np.linalg.norm(mu)
mu *= np.sqrt(covZ[0,0]) * 40
n, rho, threshold = 400, .95, 2.5
Z, covZ = AR1(n, rho)
Z += mu
threshold = 4
```

```python
f = plt.figure(figsize=(12,7))
ax = f.gca()
ax.plot(Z, color='gray', linewidth=2)
ax.plot(mu, '--', color='red', linewidth=2)
screenZ = positive_screen(Z, covZ, threshold * np.sqrt(covZ[0,0]))
screenZ.selected.sum()
```

```python
ax = f.gca()
for i, eta, center, interval in screenZ.intervals:
    lower, upper = interval
    truth = (eta * mu).sum()
    if plot:
        ax.scatter(i, truth, s=30)
    if truth > lower and truth < upper:
        covered = True
    else:
        covered = False
    ylim = ax.get_ylim()
    if plot:
        if not covered:
            ax.plot([i,i], [lower, upper], linewidth=2, color='red', alpha=0.2)
        else:
            ax.plot([i,i], [lower, upper], linewidth=2, color='green', alpha=0.2)
    ax.set_ylim(ylim)
ax.set_xlim([-1,n+1])
f
```
